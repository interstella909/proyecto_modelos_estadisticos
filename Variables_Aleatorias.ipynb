# Variables Aleatorias

Una variable aleatoria (V.A.) es una función que asigna un valor numérico a cada resultado de un experimento aleatorio.

---

## 1. Variables Aleatorias Discretas

Toman valores finitos o numerables.

### Función de probabilidad (pmf)

Ejemplo: Lanzar un dado justo

$$P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}$$

### Esperanza y Varianza

- Esperanza: $$\mathbb{E}[X] = \sum_{x} x \cdot P(X = x)$$
- Varianza: $$\mathrm{Var}(X) = \mathbb{E}[X^2] - \mathbb{E}[X]^2$$

## 2. Variables Aleatorias Continuas

Toman cualquier valor en un intervalo.

### Función de densidad (pdf)

Ejemplo: distribución uniforme en [0,1]

$$
f(x) = \begin{cases}
1, & 0 \leq x \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

### Esperanza y Varianza

- Esperanza: $$\mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$$
- Varianza: $$\mathrm{Var}(X) = \mathbb{E}[X^2] - \mathbb{E}[X]^2$$

## 4. Función Generadora de Momentos (FGM)

La Función Generadora de Momentos (FGM) de una variable aleatoria \( X \) se define como:

$$M_X(t) = \mathbb{E}[e^{tX}]$$

Sirve para obtener los momentos de la variable aleatoria. Sus propiedades incluyen:

- $$M_X^{(n)}(0) = \mathbb{E}[X^n]$$
- $$M_X'(0) = \mathbb{E}[X]$$

**Ejemplo:**  
Si $$\ X \sim \text{Bernoulli}(p) \$$, entonces:

$$M_X(t) = \mathbb{E}[e^{tX}] = (1 - p) + p e^t$$


## 5. Función Característica (FC)

La función característica de una variable aleatoria \( X \) es:

$$\varphi_X(t) = \mathbb{E}[e^{itX}]$$

donde $$\ i = \sqrt{-1} \$$ es la unidad imaginaria.

**Propiedades:**

- Siempre existe (incluso si la FGM no).
- Determina completamente la distribución de \( X \).
- Se usa frecuentemente para demostrar el **Teorema Central del Límite (TCL)**.

## 6. Ley de los Grandes Números (LGN)

Afirma que la **media muestral** de una sucesión de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) converge a la esperanza:

$$\lim_{n \to \infty} \bar{X}_n = \mathbb{E}[X]$$

### Ejemplo (Simulación en Python)

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación LGN con dados
n = 10000
dados = np.random.randint(1, 7, size=n)
media_muestral = np.cumsum(dados) / np.arange(1, n + 1)

plt.plot(media_muestral, label='Media muestral')
plt.axhline(3.5, color='red', linestyle='--', label='Esperanza teórica (3.5)')
plt.title('Ley de los Grandes Números')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Media')
plt.legend()
plt.grid(True)
plt.show()
```

## 7. Teorema Central del Límite (TCL)

Sea \( X_1, X_2, \ldots, X_n \) un conjunto de variables aleatorias i.i.d. con media \( \mu \) y varianza \( \sigma^2 \). Entonces:

$$Z_n = \frac{\sum_{i=1}^n X_i - n\mu}{\sqrt{n}\sigma} \xrightarrow{d} \mathcal{N}(0,1)$$

Esto quiere decir que la suma (o promedio) de muchas variables aleatorias independientes se aproxima a una distribución normal.

### Ejemplo (Simulación en Python)


```python
# TCL con suma de uniformes
n_experimentos = 10000
n_sumas = 30  # Número de variables sumadas

suma_vars = np.sum(np.random.uniform(0, 1, size=(n_experimentos, n_sumas)), axis=1)

plt.hist(suma_vars, bins=50, density=True, color='lightgreen', edgecolor='black')
plt.title('Teorema Central del Límite (suma de Uniformes)')
plt.xlabel('Suma de variables')
plt.grid(True)
plt.show()
```
